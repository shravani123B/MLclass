## 34.1.8 Evaluation Metrics for Classification

| Metric      | Meaning                                                                 | Ideal Value |
|------------|-------------------------------------------------------------------------|-------------|
| **Accuracy** | Percentage of correct predictions                                       | Closer to **1** |
| **Precision** | Proportion of predicted positives that are actually positive           | High |
| **Recall** | Proportion of actual positives that are correctly identified             | High |
| **F1-Score** | Harmonic mean of precision and recall (balances both)                  | High |
| **ROC-AUC** | Area under the Receiver Operating Characteristic curve                  | Closer to **1** |
| **Log Loss** | Error based on predicted probabilities                                  | Closer to **0** |

# 34.5 Separate Features and Labels

```python

# Separate features and targets
# Drop the target column "Churn from the main dataframe
fdata = data.drop(columns=['Churn'])
X = fdata
# Create a separate target variable (y)
#Encode target variable to numeric (Yes=1, No=0)
y = data["Churn"].map({'Yes':1, 'No':0})
#Identify Categorical and Numerical Features
categorical = fdata.select_dtypes(include='object').columns
numerical = fdata.select_dtypes (exclude='object').columns
# Display Feature Types
print("Categorical Features:", list(categorical))
print("Numerical Features:", list(numerical))
```
